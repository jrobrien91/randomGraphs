```python
import satpy
import s3fs
from datetime import datetime
```

## Check the Amazon S3 bucket for latest files

```python
# Calculate Day of Year to grab data from S3 bucket
DAY_OF_YEAR = datetime.now().timetuple().tm_yday
# Calculate current time in UTC
UTC_TIME = datetime.utcnow()
```

```python
UTC_TIME.hour
```

```python
print(DAY_OF_YEAR, UTC_TIME)
```

```python
# use anonymous credentials to access public data
fs = s3fs.S3FileSystem(anon=True)

# Search the s3 bucket for available files
files = fs.ls("s3://noaa-goes16/ABI-L2-CMIPC/2023/" + str(DAY_OF_YEAR) + '/' + str(UTC_TIME.hour))

# get the files
fs.get(files[-1], 'GOES16/')
```
