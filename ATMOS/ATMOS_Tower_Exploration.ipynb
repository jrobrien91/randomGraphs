```python
nlist = []
with open('/Users/jrobrien/ANL/ATMOS/data/ATMOS_2022_all_non_QC_data.txt', 'r') as fd:
    reader = fd.readlines()
    for row in reader[5:-2]:
        print(row.strip().split(' ')[0])
        if row.strip().split(' ')[0] == '283':
            tmp = [x for x in row.strip().split(' ') if len(x) > 0]
            #tmp_B = [float(x) for x in tmp[:-2] if x > 2]
            nlist.append(tmp[:-2])
```

```python
var_list = [x for x in reader[-1].strip().split(' ') if len(x) > 0]
```

```python
df = pd.DataFrame(nlist, columns = var_list[:-2])
```

```python
df
```

```python
df['T_LST'] = pd.to_datetime(df['T_LST']).apply(lambda x: x.replace(day=10))
```

```python
df
```

```python
ds = df.to_xarray()
```

```python
for var in ds.data_vars:
    if var != 'T_LST':
        ds[var].data = ds[var].data.astype(float)
```

```python
ds
```

```python
ds = ds.assign_coords({'time' : ds.T_LST})
```

```python
fig, axarr = plt.subplots(2, 2, figsize=[12,8])
plt.subplots_adjust(hspace=0.4)

ds.TaC_60m.plot(x='time', ax=axarr[0, 0], label='60m Air Temp')
ds.TaC_10m.plot(x='time', ax=axarr[0, 0], label='10m Air Temp')
ds.Tdp_10m.plot(x='time', ax=axarr[0, 0], label='10m Dept Temp')
axarr[0, 0].set_ylabel('Temperature [C]')
axarr[0, 0].set_xlabel('Time [MM-DD HH]')
axarr[0, 0].set_ylim([0, 25])
axarr[0, 0].grid(True)
axarr[0, 0].legend(loc='upper left')

ds.spdV60m.plot(x='time', ax=axarr[0, 1], label='60m Wind Speed')
ds.spdV10m.plot(x='time', ax=axarr[0, 1], label='10m Wind Spped')
axarr[0, 1].set_ylabel('Wind Speed [m/s]')
axarr[0, 1].set_xlabel('Time [MM-DD HH]')
axarr[0, 1].set_ylim([0,10])
axarr[0, 1].grid(True)
axarr[0, 1].legend(loc='upper left')

ds.dirV60m.plot(x='time', ax=axarr[1, 0], label='60m Wind Direction')
ds.dirV10m.plot(x='time', ax=axarr[1, 0], label='10m Wind Direction')
axarr[1, 0].set_ylabel('Wind Speed [m/s]')
axarr[1, 0].set_xlabel('Time [MM-DD HH]')
#axarr[0, 1].set_ylim([0,10])
axarr[1, 0].grid(True)
axarr[1, 0].legend(loc='upper left')

ds.baroKPa.plot(x='time', ax=axarr[1, 1], label='Barometric')
#ds.e__10m.plot(x='time', ax=axarr[1, 1], label='Vapor')
axarr[1, 1].set_ylabel('Pressure [kPa]')
axarr[1, 1].set_xlabel('Time [MM-DD HH]')
#axarr[0, 1].set_ylim([0,10])
axarr[1, 1].grid(True)
axarr[1, 1].legend(loc='upper left')

plt.suptitle('ATMOS 60m Tower - 10 October 2022 - All Times CST')
```
